**0. Install Dependancies**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
```

**1. Import Data Using .csv**

```python
df = pd.read_csv('TableSet.csv')
df.drop(columns=['Unnamed: 0',"ROUTE"],inplace=True)
```

**2. Define Label and Feature**

```python
y = df['CRASH_SEVERITY_ID'] # Label
X = df.drop(columns=['CRASH_SEVERITY_ID']) # Features
X
```

**3. Install Dependancies**

WOW!! and adjusted R-squared of .803! We were really shocked by this result especially since this was just a plug and run. A few things to note

> - This model used values like hour and city_id which are values that aren't continuous. 
> - Despite the result being the highest, our model was run without being centered (without a y intercept). Now even though our label is skewed we can't evaluate it as continuous because we represent severity in this case as as grade rather than quantative value. There is not much we can do to correct that.
>Lastly we are notified that we have a massive condition number notifying us that probably have multicollinearity
