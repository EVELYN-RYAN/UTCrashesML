**0. Setup**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

df = pd.read_csv('TableSet.csv')
df.drop(columns=['Unnamed: 0',"ROUTE"],inplace=True)
```

**1. Correct Categorical Features**

```python
df['HOUR'] = df['HOUR'].astype(str)
df['WEEKDAY'] = df['WEEKDAY'].astype(str)
df['MONTH'] = df['MONTH'].astype(str)
df.dtypes
```

**2. Dummie Code Categories**

```python
for col in df:
          if not pd.api.types.is_numeric_dtype(df[col]) and col != 'CRASH_SEVERITY_ID':
            df = pd.get_dummies(df, columns=[col])
df.head()
```

```python
y = df['CRASH_SEVERITY_ID'] # Label
X = df.drop(columns=['CRASH_SEVERITY_ID','CITY_ID','COUNTY_ID']) # Features
X
```

**4. Fit and Evaluate**

As you can see we, after we had correctly gone about treating our categorical values properly we get nothing even close to .803. I will leave all additional evaluation behind

```python
model = sm.OLS(y, X).fit()

# View results
print(model.summary())
```
